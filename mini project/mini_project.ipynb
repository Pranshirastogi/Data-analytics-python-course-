```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('dataset.csv')
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
mf = df['female_percentage'].astype('float').mean()
mm = df['male_percentage'].astype('float').mean()
mn = df['num_students'].astype('float').mean()
ms = df['student_staff_ratio'].astype('float').mean()
```

```python
df['female_percentage'].replace(np.nan,mf,inplace=True)
df.head()
```

```python
df['male_percentage'].replace(np.nan,mm,inplace=True)
df['num_students'].replace(np.nan,mn,inplace=True)
df['student_staff_ratio'].replace(np.nan,ms,inplace=True)
df.head()
```

```python
mi = df['international_students'].astype('float').mean()
df['international_students'].replace(np.nan,mi,inplace=True)
df.head()
```

```python
my = df['income'].astype('float').mean()
df['income'].replace(np.nan,my,inplace=True)
df.head()
```
