col related operations

```python
df.rename(mapper={'make':'brand'},axis=1,inplace=True) # renaming index 'mapper'
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.head()
```

```python
df.replace('alfa-romero','alfa-romeo',inplace=True)
```

```python
df.replace('?',np.nan,inplace=True)  # remove all ? by np.nan
```

```python
df
```

```python
# df.dropna() removes row of NaN
```

```python
df.dropna(subset=['price'],inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()  # count the no. of missing data in each column
```

```python
df['normalized-losses'].astype('float')
```

```python
df['normalized-losses'].astype('float').sum()
```

```python
mnl = df['normalized-losses'].astype('float').mean()
```

```python
df['normalized-losses'].replace(np.nan,mnl,inplace=True)
df.head()
```

```python
mb = df['bore'].astype('float').mean()
```

```python
df['bore'].replace(np.nan,mb,inplace=True)
df.head()
```

```python
df
```

```python
ms = df['stroke'].astype('float').mean()
```

```python
df['stroke'].replace(np.nan,ms,inplace=True)
```

```python
df['horsepower'].replace(np.nan,df['horsepower'].astype('float').mean(),inplace=True)
df['peak-rpm'].replace(np.nan,df['peak-rpm'].astype('float').mean(),inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].replace(np.nan,'four',inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv('automobile_cleaned.csv',index=False)
```
