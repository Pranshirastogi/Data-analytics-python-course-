```python
def fix_relation_value(code):
    if code == 101:
        return 'head'
    elif code == 201:
        return 'spouse'
    elif code == 301:
        return 'child'
    elif code == 303:
        return 'stepchild'
    elif code == 501:
        return 'parent'
    elif code == 701:
        return 'sibling'
    elif code == 901:
        return 'grandchild'
    elif code == 1001:
        return 'other'
    elif code == 1113:
        return 'partner'
    elif code == 1115:
        return 'housemate'
    elif code == 1241:
        return 'roomer'
    elif code == 1242:
        return 'foster'
    elif code == 1260:
        return 'non-relate'
```

```python
df['relate'].apply(fix_relation_value) #testing
```

```python
df['relate']=df['relate'].apply(fix_relation_value)
```

```python
def fix_maritalstat_value(code):
    if code == 1:
        return 'present spouse'
    elif code == 2:
        return 'absent spouse'
    elif code == 3:
        return 'seprated'
    elif code == 4:
        return 'divorced'
    elif code == 5:
        return 'widowed'
    elif code == 6:
        return 'never married'
```

```python
df['marst']=df['marst'].apply(fix_maritalstat_value)
```

```python
def fix_employment_value(code):
    if code == 10:
        return 'employed'
    elif code == 12:
        return 'unemployed'
```

```python
df['empstat']=df['empstat'].apply(fix_employment_value)
```
