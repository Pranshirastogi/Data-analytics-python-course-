```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [21,345,56,12,5,47,1]
plt.plot(y)
```

```python
plt.figure(figsize=(15,4))
plt.pie(y)
plt.show() # to hide unnecessary data
```

```python
fig , ax=plt.subplots(nrows=1, ncols=2)
plt.show()
```

```python
fig , ax=plt.subplots(nrows=1, ncols=2 , figsize=(15,2))
plt.show()
```

```python
t = np.arange(-3*np.pi , 3*np.pi, 0.2)
x = np.sin(t)
y = np.cos(t)
fig, ax=plt.subplots(nrows=1,ncols=2,figsize=(15,2))
ax[0].plot(t,x,'r')
ax[1].bar(t,y)
plt.show()
```

```python
x = [100,200,300,400,500,600,700,800,900,1000]
y = [23,45,56,67,78,89,99,90,30,45]
plt.plot(x,y,linewidth=2,color='purple', marker='h', markersize=15 , markerfacecolor='white',markeredgecolor='green',linestyle='--')
```

```python
plt.plot(x,y, 'go--',linewidth=2, markersize=15 , markerfacecolor='white',markeredgecolor='green')
plt.show()
```

```python
plt.plot(x,y, 'go--',linewidth=2, markersize=15 , markerfacecolor='white',markeredgecolor='green')
plt.title('Some graph showing somthing',fontdict={'fontsize':10,'fontweight':'light','color':'red','fontfamily':'monospace'})
plt.xlabel('X axis values')
plt.ylabel('Y axis values', fontdict={'color':'red'})
plt.text(800,60,'this is serious')
plt.annotate("Peak value",xy=(400,68),xytext=(400,30),arrowprops=dict(facecolor='black',shrink=0.05))
plt.show()
```
